```python
import matplotlib.pyplot as plt

# 데이터 준비 (이미지에서 직접 값을 추출하여 입력)
year_2019 = {'Gasoline': 75, 'Diesel': 10, 'HEV': 5, 'PHEV': 2, 'MHEV': 8, 'BEV': 0}
year_2030 = {'Gasoline': 40, 'Diesel': 5, 'HEV': 3, 'PHEV': 4, 'MHEV': 10, 'BEV': 38}

# 데이터 프레임 생성 (pandas를 사용하면 더욱 효율적으로 데이터를 관리할 수 있습니다)
import pandas as pd
data = {'2019': year_2019, '2030': year_2030}
df = pd.DataFrame(data)

# 그래프 그리기
df.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.ylabel('Share of car sales')
plt.title('Global Car Sales by Fuel Type (2019 vs. 2030)')
plt.legend(loc='lower center', ncol=3)
plt.show()
```
